```python
import numpy as np
import pandas as pd
```

```python
orginal_data = pd.read_csv("original_data_104.csv")
salary = orginal_data["薪資待遇"]
for word in salary:
    new_salary=""
    for char in word:
        if char.isdigit() or char=="~": #isdigit:可判斷字串內是否為數字,是回傳True,否回傳False
            new_salary+=char #幾個char就串幾個字
            #用"~"區隔並找出薪資上下限
    if "~" in new_salary:
        h1=new_salary[:new_salary.find("~")] #find:對字串內尋找指定字元,並回傳所在第幾位
        h2=new_salary[new_salary.find("~")+1:]
    else: #ex:月薪32,000元以上
        h1=new_salary #讓上下限都是同樣數字
        h2=new_salary
    # h1=int(h1)
    # h2=int(h2)
    print(h1,h2)
            #產出g&h1&h2作為output

     #筆數間的分隔線
        
    #抓取並解析網頁資料
    #屬性順序並不會影響本體,因此為操作方便,把"&page=2"剪下並移至末端"
```
